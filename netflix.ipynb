### convert the seasons into minutes  and replace minute with blank spaces

```python
def convert_to_minutes(x):
    if pd.isna(x):
        return np.nan

    x=x.lower().strip()

    if 'min'in x:
        return int(x.replace('min',"").strip())

    if 'Season'in x:
        Seasons=int(x.split()[0])
        return Seasons*450

    return np.nan

df['duration_minutes']=df['duration'].apply(convert_to_minutes)
    
    
    
```

```python
df['duration_minutes']=df['duration_minutes'].fillna(df['duration_minutes'].mean())
```

## feature enginerring

```python
features=["type","director","cast","country","listed_in","description","release_year","duration_minutes"]
df_enc=df[features].fillna("")
```

### convert the text to numerical vector

```python
from sklearn.feature_extraction.text import TfidfVectorizer
text_features=(
    df_enc['director']+''+
    df_enc['cast']+''+
    df_enc['country']+''+
    df_enc['listed_in']+''+
    df_enc['description']
)

tfidf=TfidfVectorizer(stop_words="english",max_features=5000)
text_encoded=tfidf.fit_transform(text_features)
```

```python
from sklearn.preprocessing import OneHotEncoder
encoder=OneHotEncoder()
type_encoded=encoder.fit_transform(df_enc[['type']])
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()

numeric_encoded=scaler.fit_transform(df_enc[['release_year','duration_minutes']])
```

```python
from scipy.sparse import hstack
final_features=hstack([text_encoded,type_encoded,numeric_encoded])
```

```python
final_features
```

```python
final_features=final_features.tocsr()
```

### KNN  model training

```python
from sklearn.neighbors import NearestNeighbors
knn_model=NearestNeighbors(
    n_neighbors=6,
    metric='cosine',
    algorithm='brute'
    )
knn_model.fit(final_features)
```

## create a recommendation function

```python
def recommend_show(title,n=5):
    idx=df[df['title']==title].index[0]

    distances,indices=knn_model.kneighbors(

        final_features[idx:idx+1],
        n_neighbors=n+1
    )

    recommended_indices=indices[0][1:]

    return df.iloc[recommended_indices][['title','listed_in','cast']

    ]
    
    
```

```python
recommend_show("Baazaar")
```
